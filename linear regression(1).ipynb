```python
# Regresssion: It is basically a ststistical approach to find the correlation between the variables.(Dependent and independent).Regression algorithm gives us a continuous output, which means if we are asked to built a model which predicts in future outcomes where the output will be continuous.It is a supervised learning algorithm.

# Linear Regression: It is a ststistical method which is used for predictive analysis.Linear regression makes predictions for continuous/ real numeric variables such as cost, age, sales, temperature, product price etc.
```

```python
#Q: Find out the permium paid by the persons whose ages are 21 and 50 from the term Insurence dataset using linear regression algorithm.
```

```python
#Step-1:Import perticular libraries for analysis as well as predictions.
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import linear_model
```

```python
# Step-2: to read our dataset we are using pandas function
a = pd.read_csv("Term Insurence.csv")
# Again to view our dataset we have
print("Our dataset becomes:")
a
```

```python
# Here the are two variables are there such as  dependent variabe and the other one is the independent variable, where x-axis repersents the independent variable while y-axis represents the dependent variable.
# Now our aim is to plot the dependent as well as independent variable
# So, to view our dataset we have
sns.lmplot(x="age",y="Premium",data=a)
# Here we got the best-fit line along with the plot.
```

```python
#Import the linear regression method from the linear model
reg = linear_model.LinearRegression()
```

```python
# Now we have to train our model with the datapoints, so we are using fit method.Here we have to provide the two dimensional parameter inside the fit() method.
reg.fit(a[['age']],a['Premium'])
```

```python
# Now our model is trained it is ready for Prediction.So now we have to find out the premium paid by the person whose age is 21
reg.predict([[21]]) 



```

```python
# Premium paid by the person whose age is 50
reg.predict([[50]]) 
```

```python
# Now we have to validate our result using the linear equation y=m*x+c
# For that we have to find the slope
reg.coef_
```

```python
# Now we have to find the intercept
reg.intercept_
```
